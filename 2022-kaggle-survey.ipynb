```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go

colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']
kaggle_colors = ['#39c5ff', '#ffffff', '#f2f2f2', '#9ca3a4', '#3f484b']
font = 'Rubik'
```

```python
df = pd.read_csv('../input/kaggle-survey-2022/kaggle_survey_2022_responses.csv')
```

```python
df = df[1:]
# df['Duration (in seconds)'] = df['Duration (in seconds)'].astype('int')
```

```python
df.loc[df["Q3"] ==  "Nonbinary"] = "Others"
df.loc[df["Q3"] == "Prefer not to say"] = "Others"
df.loc[df["Q3"] == "Prefer to self-describe"] = "Others"
df["Q3"].unique()
```

```python
df.loc[df["Q4"] == "Other"]="Others"
```

```python
gender_count = df.groupby(['Q3']).size().reset_index().rename(columns={0:'count'})
fig = px.bar(data_frame=gender_count,x='Q3',y='count',color='Q3',title="Gender Count",text_auto=True)
fig.show()
```

```python
duration_count = df.groupby(['Duration (in seconds)']).size().reset_index().rename(columns={0:'count'})
duration = duration_count[:1000]
fig = px.histogram(data_frame=duration,x='Duration (in seconds)',y='count',title="Duration Histogram",text_auto=True)
fig.show()
```

```python
age_count = df.groupby(['Q2']).size().reset_index().rename(columns={0:'count'})
fig = px.bar(data_frame=age_count, x='Q2', y='count',color='Q2',template='plotly_dark',color_discrete_sequence=['#39c5ff','#F5F5F1'],
                 text_auto=True,barmode='stack',
          title="Age Countplot")
fig.update_traces(textfont_size=20, textangle=0, textposition="outside", cliponaxis=False)

fig.update_layout(
    title="Age Countplot",
    xaxis_title="Age Groups",
    yaxis_title="Count",
    legend_title="Age Groups",
    font=dict(
        family="Rubik",
        size=12
    )
)

fig.show()
```

```python
country_count = df.groupby(['Q4']).size().reset_index().rename(columns={0:'count'})
country_count = country_count.sort_values(by=['count'],ascending=False).reset_index(drop=True)
country_count = country_count[:10]
country_count
fig = px.bar(data_frame=country_count,x='Q4',y='count',title='Country Count',text_auto=True,color='Q4')
fig.show()
```

```python
country_df = df.groupby(['Q4']).size().reset_index().rename(columns={0:'count'})
fig = px.choropleth(country_df,locations='Q4',color="count",locationmode='country names',range_color=[0,5000])
fig.update_layout(
    title="World Map",
    font=dict(
        family="Rubik",
        size=14
    )
)
fig.show()
```

```python
student_count = df.groupby(['Q5']).size().reset_index().rename(columns={0:'count'})
fig = go.Figure(data=[go.Pie(labels=student_count['Q5'],values=student_count['count'],hole=.4)])
fig.show()
```

```python
degree_count = df.groupby(['Q8']).size().reset_index().rename(columns={0:'count'})
degree_count.loc[degree_count['Q8'] == "Some college/university study without earning a bachelor’s degree","Q8"]= "College Without Bachelors"
fig = go.Figure(go.Pie(labels=degree_count["Q8"],values=degree_count["count"],hole=.4,hoverinfo='label+percent',textinfo='value',title="Degree Count"))
fig.show()
```

```python
experience_count = df.groupby(['Q16']).size().reset_index().rename(columns={0:'count'})
fig = go.Figure(go.Pie(labels=experience_count['Q16'],values=experience_count['count'],hole=.4,textinfo='value',title="Ml Experience"))
fig.show()
```

```python
annual_income_df = df.groupby(['Q29']).size().reset_index().rename(columns={0:'count'})
annual_income_df = annual_income_df.sort_values(by='count',ascending=False).reset_index(drop=True)
fig = go.Figure(go.Pie(labels=annual_income_df['Q29'],values=annual_income_df['count'],textinfo='value',hole=.4))
fig.show()
```

```python
gender_duration_count = df.groupby(['Duration (in seconds)', 'Q3']).size().reset_index().rename(columns={0:'count'})
gender_duration_count = gender_duration_count[:3000]
fig = px.histogram(data_frame=gender_duration_count,x="Duration (in seconds)",y="count",marginal="box",color='Q3')
fig.show()
```

```python
country_gender = df.groupby(['Q4','Q3']).size().reset_index().rename(columns={0:'count'})

fig = px.treemap(data_frame=country_gender,values='count',path=['Q4','Q3'])
fig.show()
```

```python
df[df['Duration (in seconds)'] == "Others"] = 0
df['Duration (in seconds)'] = df['Duration (in seconds)'].astype('float')
age_duration_df = df[df['Duration (in seconds)'] < 2000].groupby(['Q2'])['Duration (in seconds)'].mean().round(1).reset_index()
age_duration_df
```

```python
education_gender = df.groupby(['Q8','Q3']).size().reset_index().rename(columns={0:'count'})
fig = px.treemap(education_gender,path=['Q8','Q3'],values='count',title='Education X Gender Tree Map')
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

```python
top_9_countries=['United States of America','Other', 'Brazil',
                   'Nigeria', 'Pakistan', 'Japan', 'China', 'Egypt', 'Mexico']
top_9_countries_df = df[df['Q4'].isin(top_9_countries)]
fig = px.density_heatmap(top_9_countries_df,x='Q4',y='Q2',marginal_x='histogram',marginal_y='histogram',histfunc='count')
fig.show()
```

```python
gender_student = df.groupby(['Q3','Q5']).size().reset_index().rename(columns={0:'count'})

fig = px.sunburst(gender_student,values='count',path=['Q3','Q5'],title="Gender X Student Sunburst")
fig.update_traces(textinfo='label+percent parent')
fig.show()
```
